####1st Question


```python
import pandas as pd
vegetables = pd.read_csv("main.csv")
vegetables.head()
```

```python
vegetables.groupby('product_description')['price'].mean()
```

```python
vegetables["price_new"] = vegetables['price'].fillna(
   vegetables.groupby('product_description')['price'].transform("mean")
)
```

```python
vegetables[vegetables["price"].isna()].head()
```

```python
vegetables.to_csv('question1_output',index=False)
```

####2nd question

```python
vegetables["sales_amount"] = vegetables["sales_quantity"].where(
    vegetables["unit"] != "pcs",
    vegetables["product_description"].str.split("-", expand=True)[1].astype("float") * \
    vegetables["sales_quantity"]
    )
```
